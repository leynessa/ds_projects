# **Project : LUI**

This project for the displine LUI for my MBA Bimaster which aims to extract information from a chosen data source, utilizing either an API or web scraping technique. The primary goal is to demonstrate the practical application of concepts learned throughout the course

For this project, I have choosen to utilize the OpenWeather API to obtain meteorological information from various cities around the world. The selection of the OpenWeather API was based on its ease of access and the richness of information provided, allowing us to explore different aspects of meteorological data.

Throughout the project, I will demonstrate the process of data extraction and storage of this information in a DataFrame, and subsequently, perform sorting and filtering operations to analyze and visualize the data more effectively.


source: API - https://openweathermap.org/api)

```python
import requests
import pandas as pd
api_key ="f9468742a4fc6276b483de5a6094dfcf"
# List of cities
cities = ['New York', 'kingston', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose','Cairo','Cape Town','Caracas','Casablanca', 'Cayenne', 'Makassar', 'Naples','Riga','Rio de Janeiro','Rome','Run']

# Initialize lists to store data
city_names = []
countries = []
temperatures = []
feels_likes = []
weather_descriptions = []
longitudes = []
latitudes = []
humidities = []
wind_speeds = []



```

```python
# the temp is given in kelvin
def kelvin_to_celsius (kelvin):
  celsius  = kelvin-273.15
  return celsius
```

```python
# Loop through each city and get weather data
for city_name in cities:
    # Make the API request
    response = requests.get(f"https://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={api_key}")
    response = response.json()


    # Extract relevant information from the response
    country = response['sys']['country']
    temp_celsius =  response['main']['temp']
    feels_like_celsius =  response['main']['feels_like']
    weather_description =  response['weather'][0]['description']
    longitude =  response['coord']['lon']
    latitude =  response['coord']['lat']
    humidity =  response['main']['humidity']
    wind_speed =  response['wind']['speed']

    # Append data to lists
    city_names.append(city_name)
    countries.append(country)
    temperatures.append(temp_celsius)
    feels_likes.append(feels_like_celsius)
    weather_descriptions.append(weather_description)
    longitudes.append(longitude)
    latitudes.append(latitude)
    humidities.append(humidity)
    wind_speeds.append(wind_speed)

# Create DataFrame
data_dict = {
    "City": city_names,
    "Country": countries,
    "Temperature (°C)": temperatures,
    "Feels Like (°C)": feels_likes,
    "Weather Description": weather_descriptions,
    "Longitude": longitudes,
    "Latitude": latitudes,
    "Humidity (%)": humidities,
    "Wind Speed (m/s)": wind_speeds
}
df = pd.DataFrame(data_dict)

# Print DataFrame
print(df)

```
